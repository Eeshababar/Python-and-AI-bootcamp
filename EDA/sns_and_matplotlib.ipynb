```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Matplotlib

Purpose of Matplotlib

Foundational plotting library in Python.

Provides low-level control over every element of a plot (axes, labels, ticks, colors, etc.).

Flexible but often verbose — you have to write more code to get polished visuals.

Think of it as the “engine” for plots

#Seaborn



Built on top of Matplotlib.

Provides high-level functions for creating beautiful and statistically meaningful plots with less code.

Integrates tightly with pandas DataFrames, so you can pass data directly.

Has default themes and color palettes that look cleaner without extra styling.

Especially useful for statistical plots (correlation heatmaps, distribution plots, categorical plots).

You can modify the dataset, labels, colors, and other plot details to suit your specific needs.

This line of code generates a histogram of the sepal_length column from the df DataFrame.

plt.hist() is the function from the matplotlib.pyplot library used to create histograms.
df['sepal_length'] specifies the data to be used for the histogram.
bins=10 sets the number of bins (intervals) for the histogram to 10.
color='blue' sets the color of the histogram bars to blue.
alpha=0.7 sets the transparency of the bars to 0.7 (1.0 is fully opaque, 0.0 is fully transparent).

```python
# Load Iris dataset
df = sns.load_dataset('iris')
# Histogram of sepal_length from Iris dataset
plt.hist(df['sepal_length'], bins=10, color='blue', alpha=0.5)
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot of all numeric columns in the Iris dataset
sns.boxplot(data=df)
plt.title('Box Plot of Iris Features')
plt.show()
```

This line of code generates a scatter plot using Seaborn:

sns.scatterplot() is the function from the seaborn library used to create scatter plots.
data=df specifies that the data for the plot comes from the df DataFrame.
x='sepal_length' sets the 'sepal_length' column as the data for the x-axis.
y='petal_length' sets the 'petal_length' column as the data for the y-axis.
hue='species' colors the points based on the 'species' column, allowing you to distinguish between different species.
palette='Set1' specifies the color palette to be used for coloring the points based on species.
s=100 sets the size of the markers (points) in the scatter plot to 100

```python


# Scatter plot with different colors for each species
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species', palette='Set1', s=100)

# Adding title and labels
plt.title('Scatter Plot: Sepal Length vs Petal Length', fontsize=14)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')

# Show the plot
plt.show()
```

```python
# Line chart of Sepal Length vs index (for visualization)
plt.plot(df['sepal_length'], label='Sepal Length')
plt.plot(df['petal_length'], label='Petal Length')
plt.title('Line Chart of Sepal Length and Petal Length')
plt.xlabel('Index')
plt.ylabel('Length (cm)')
plt.legend()
plt.show()
```

plt.pie() is the function from the matplotlib.pyplot library used to create pie charts.
species_count provides the data for the pie chart. In this case, it's a pandas Series containing the count of each species.
labels=species_count.index sets the labels for each slice of the pie chart using the index of the species_count Series (which are the species names).

```python
# Pie chart for species distribution
species_count = df['species'].value_counts()
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Species Distribution')
plt.show()
```

```python
# Donut chart for species distribution
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', startangle=140, wedgeprops={'width': 0.4})
plt.title('Donut Chart for Species Distribution')
plt.show()
```
