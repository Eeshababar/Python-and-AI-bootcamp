Outlier detection

The selected code defines a function iqr_outliers that identifies outliers in a pandas Series using the Interquartile Range (IQR) method.

Here's a breakdown:

import pandas as pd: This line imports the pandas library and aliases it as pd, which is a standard convention.
def iqr_outliers(series: pd.Series, k: float = 1.5):: This defines a function named iqr_outliers that takes two arguments:
series: A pandas Series (annotated with : pd.Series for type hinting).
k: A float value with a default of 1.5. This is a multiplier used in the IQR calculation to determine the bounds for outlier detection.
Q1 = series.quantile(0.25): Calculates the first quartile (25th percentile) of the input series.
Q3 = series.quantile(0.75): Calculates the third quartile (75th percentile) of the input series.
IQR = Q3 - Q1: Calculates the Interquartile Range (IQR), which is the difference between the third and first quartiles.
lower, upper = Q1 - k*IQR, Q3 + k*IQR: Determines the lower and upper bounds for outlier detection. Any value below the lower bound or above the upper bound is considered an outlier. The k value (defaulting to 1.5) controls how far from the quartiles the bounds are set.
return series[(series < lower) | (series > upper)]: This line filters the original series and returns only the values that are less than the lower bound or greater than the upper bound. These are the identified outliers.


```python
import pandas as pd

def iqr_outliers(series: pd.Series, k: float = 1.5):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower, upper = Q1 - k*IQR, Q3 + k*IQR
    return series[(series < lower) | (series > upper)]

# Example
data = pd.Series([10,12,12,13,12,11,300,12,11,10])
print(iqr_outliers(data))
```

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
from scipy import stats

# Step 2: Create example data
data = pd.Series([10, 12, 11, 13, 12, 11, 300, 12, 10, 11])

# Step 3: Calculate Z-scores
z_scores = np.abs(stats.zscore(data))

# Step 4: Identify outliers (threshold = 3)
outliers = data[z_scores > 3]

# Step 5: Print results
print("Z-scores:\n", z_scores)
print("\nDetected Outliers:\n", outliers)
```

Label Encoding is a common technique for converting categorical variables into numerical values. Each unique category value is assigned a unique integer based on alphabetical or numerical ordering.

```python
from sklearn.preprocessing import LabelEncoder

data = ["Red", "Green", "Blue", "Red"]
encoder = LabelEncoder()
encoded = encoder.fit_transform(data)

print(encoded)  # [2 1 0 2]
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

data = np.array([["Red"], ["Green"], ["Blue"], ["Red"]])
encoder = OneHotEncoder()
encoded = encoder.fit_transform(data).toarray()

print(encoded)
```

```python
from sklearn.preprocessing import OrdinalEncoder
import numpy as np

# Example dataset
data = np.array([
    ["Low"],
    ["Medium"],
    ["High"],
    ["Low"],
    ["High"]
])

# Define the order of categories explicitly
encoder = OrdinalEncoder(categories=[["Low", "Medium", "High"]])

# Transform the data
encoded = encoder.fit_transform(data)

print(encoded)
```

✨ Beginner Shortcut:

If categories are ordered → ✅ Ordinal Encoder

If categories are unordered, small number → ✅ One-Hot Encoder

If categories are unordered, large number → ✅ Label/Target/Frequency encoding (to avoid huge datasets).
