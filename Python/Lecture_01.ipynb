## Variable Names:
A variable can have a short name (like x and y) or a more descriptive name (age, carname, total_volume).

## Rules for Python variables:

* A variable name must start with a letter or the underscore character
* A variable name cannot start with a number
* A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
* Variable names are case-sensitive (age, Age and AGE are three different variables)
* A variable name cannot be any of the Python keywords.

**Using if-elif-else (Multiple Conditions)**

```python
marks = int(input("Enter your marks: "))
#marksstr = str(marks)
if marks >= 90:
    print("Grade: A ")
elif marks >= 80:
    print("Grade: B ")
elif marks >= 70:
    print("Grade: C ")
else:
    print("Grade: Fail ")
```

Nested if Example (Important)
:
 Real-life example: If it's a holiday, check whether it's raining or sunny.

```python
is_holiday = input("Is it a holiday? (yes/no): ").lower()

if is_holiday == "yes":
    weather = input("What's the weather like? (rainy/sunny): ").lower()

    if weather == "rainy":
        print("Stay home and watch a movie ")
    elif weather == "sunny":
        print("Go out and enjoy the sun ")
    else:
        print("Enjoy your day however you like! ")
else:
    print("Get ready for work or school! ")
```
